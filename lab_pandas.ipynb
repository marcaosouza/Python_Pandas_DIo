```python
# importando pandas
import pandas as pd
```

```python
df1 = pd.read_excel('/content/Aracaju.xlsx')
df2 = pd.read_excel('/content/Fortaleza.xlsx')
df3 = pd.read_excel('/content/Natal.xlsx')
df4 = pd.read_excel('/content/Recife.xlsx')
df5 = pd.read_excel('/content/Salvador.xlsx')
```

```python
# concatenando dfs
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
df.sample(5)
```

```python
df['Vendas'].fillna(df['Vendas'].mean(), inplace=True)
```

```python
df.sample(20)
```

```python
df['Vendas'].mean()
```

```python
df.describe()
```

```python
# criando colunas - receita
df["Receita"] = df['Vendas'].mul(df["Qtde"])
```

```python
df.head()
```

```python
df['Receita'].describe()
```

```python
df.nsmallest(5,'Receita')
```

```python
df.nlargest(10, 'Receita')
```

```python
df.groupby('Cidade')['Receita'].sum()
```

```python
df.groupby('Cidade')['Receita'].describe()
```

```python
df.dtypes
```

```python
# exemplificando alterao no tipo de dados
df['Data'] = df['Data'].astype('int64')
```

```python
df.dtypes
```

```python
# retornando coluna data para datetime
df['Data'] = pd.to_datetime(df['Data'])
```

```python
df.head()
```

```python
# agrupando por ano
df.groupby(df['Data'].dt.year)['Receita'].sum()
```

```python
# inserindo nova coluna com valor (ano)
df['Ano_Venda'] = df['Data'].dt.year
```

```python
df.sample(15)
```

```python
df['Ms_Venda'], df['Dia_Venda'] = (df['Data'].dt.month, df['Data'].dt.day)
```

```python
df.sample()
```

```python
# data mais antiga
df['Data'].min()
```

```python
# dif datas
df['Diferena_Dias'] = df['Data'] - df['Data'].min()
```

```python
df.sample(10)
```

```python
# inserindo trimestre
df['Trimestre_Venda'] = df['Data'].dt.quarter
```

```python
df.sample()
```

```python
# filtro vendas maro 2019
vendas_marco = df.loc[(df['Data'].dt.year == 2019) & (df['Data'].dt.month == 3)]
```

```python
vendas_marco.sample(10)
```
